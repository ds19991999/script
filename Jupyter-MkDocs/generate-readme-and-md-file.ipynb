## 生成ipynb目录

```python
import os
import os.path
# 文件名
folders = ['1-jupyter-skills',
           '2-latex',
           '3-python-tools', 
           '4-python-essentials',
           '5-advanced-python',
           '6-numpy',
           '7-scipy',
           '8-matplotlib',
           '9-object-oriented-programming',
           '10-other-languages',
           '11-something-interesting',
           '12-pandas']
# 对应的中文名
chinese = ['一、Jupyter 技巧',
           '二、Latex 笔记',
           '三、Python 工具', 
           '四、Python 基础',
           '五、Python 进阶',
           '六、Numpy 学习',
           '七、Scipy 基础',
           '八、Matplotlib 基础',
          '九、面向对象编程',
          '十、其他语言',
          '十一、有趣的例子',
           '十二、Pandas 初探']
# 对应到字典中
folders_to_chinese = dict(zip(folders, chinese))
# 产生目录文件：
index_file = open('README.md', 'w')
# 遍历产生中文目录
import nbformat
import glob
index_file.write('# Contents \n')
print "Contents"
index_file.write('* 该分支的md笔记：[md-files](md-files/)\n\n')
for folder in folders:
    index_file.write('## [' + folders_to_chinese[folder] 
                     +']('+ 'md-files/'+ folder + ')\n')
    print folders_to_chinese[folder]
    files = sorted(os.listdir(folder))
    for file_name in files:
        if file_name.endswith('.ipynb'):
            with open(os.path.join(folder, file_name)) as fp:
                nb = nbformat.read(fp, nbformat.NO_CONVERT)
            name = nb['cells'][0]['source'][1:].strip()
            print '    ' + file_name[:2], name
            index_file.write('* [' + file_name[:2] + ' ')
            index_file.write(name.encode('utf-8'))
            index_file.write('](' + folder + '/' + file_name +')\n')
index_file.close()
```

## 遍历文件及定义输出函数

```python
import nbconvert
import glob
# 对应到字典中
folders_to_chinese = dict(zip(folders, chinese))
# 检查路径是否存在，删除旧的文件：
if not os.path.exists('md-files'):
    os.mkdir('md-files')
    
for n in glob.glob('md-files/*/*/*'):
    print n
    os.remove(n)
# 遍历文件夹得到所有的文件名：
file_names = []
for folder in folders:
    files = sorted(os.listdir(folder))
    file_names += [os.path.join(folder, file_name) for file_name in files if file_name.endswith('.ipynb')]
def convert_to_files(names, to_format):
    target_dir = os.path.join("md-files")
    for folder in folders:
        if not os.path.exists(os.path.join(target_dir, folder)):
            os.makedirs(os.path.join(target_dir, folder))
    
    for file_name in names:
        p = nbconvert.export(nbconvert.get_exporter(to_format), file_name)
        with open(os.path.join(target_dir, file_name[:-6] + p[1]["output_extension"]), 'w') as f:
            f.write(p[0].encode("utf-8"))
        print file_name
```

## 输出markdown

```python
# 输出md文件
convert_to_files(file_names, "markdown")
```
