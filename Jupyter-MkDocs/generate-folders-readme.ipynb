# 生成md目录

```python
import os
import os.path
##############################################文件名###########################################
# 文件名
folders = ['1-Guide',
           '2-Jupyter-skills',
           '3-Git',
           '4-Regular-Expression',
           '5-Python-Tools', 
           '6-Python-Essentials',
           '7-Advanced-Python',
           '8-Numpy',
           '9-Scipy',
           '10-Matplotlib',
           '11-object-oriented-programming',
           '12-other-languages',
           '13-something-interesting',
           '14-pandas',
           'Date-Structure']
# 对应的中文名
chinese = ['一、Guide 学习向导',
           '二、Jupyter 技巧',
           '三、Git 学习笔记',
           '四、Re 正则表达式',
           '五、Python 工具', 
           '六、Python 基础',
           '七、Python 进阶',
           '八、Numpy 学习',
           '九、Scipy 基础',
           '十、Matplotlib 基础',
           '十一、面向对象编程',
           '十二、其他语言',
           '十三、有趣的例子',
           '十四、Pandas 初探',
           'C 数据结构']
# 对应到字典中
folders_to_chinese = dict(zip(folders, chinese))
#####################################先产生文件夹内部README.md################################
for folder in folders:
    folder_file = open(folder+'/README.md','w')
    folder_file.write('# Contents \n')
    files = sorted(os.listdir(folder)) 
    # 不处理README.md
    i = 0
    for file_name in files:
        i += 1
        if file_name.endswith('.md') and file_name != 'README.md':          
            fname = folder+'/'+file_name
            with open(fname) as fp:
                lines = fp.readlines()
                for f in lines:
                    if f[0] == '#':
                        new_name = f
                        break   
                # print '    '+str(i)+new_name[1:-1]
            folder_file.write('- ['+ str(i)+new_name[1:-1]+']('+ file_name +')\n')    
    folder_file.close()
#####################################再产生文件夹外部README.md#################################
# 产生目录文件：
index_file = open('README.md', 'w')
index_file.write('# Contents \n')
print 'Contents'
for folder in folders:
    index_file.write('* [' + folders_to_chinese[folder]
                     # 这里的转换要注意，将大写全部转为小写，空格替换为‘-’
                     +'](#' + folders_to_chinese[folder].lower().replace('_','-').replace(' ','-').replace('、','') + ')\n')
    print '    '+folders_to_chinese[folder]  
index_file.write('\n')

for folder in folders:
    # 处理文件夹名
    index_file.write('## ' + folders_to_chinese[folder]+'\n')
    print folders_to_chinese[folder]
    files = sorted(os.listdir(folder))    
    # 单独处理README
    for file_name in files:
        if file_name == 'README.md':
            i = 1
            name = file_name
            fname = folder+'/'+name
            with open(fname) as fp:
                lines = fp.readlines()
                for f in lines:
                    if f[0] == '#':
                        new_name = f
                        break    
                print '    '+str(i)+' 目录'                  
            index_file.write('* [' + str(i)+' 目录')
            index_file.write('](' + folder + '/' + file_name +')\n')   
    # 处理文件夹内其他内容
    for file_name in files:       
        if file_name.endswith('.md') and file_name != 'README.md':
            i += 1
            name = file_name            
            fname = folder+'/'+name
            with open(fname) as fp:
                lines = fp.readlines()
                for f in lines:
                    if f[0] == '#':
                        new_name = f
                        break   
                print '    '+str(i)+new_name[1:-1]  
                
            index_file.write('* [' + str(i)+new_name[1:-1])
            index_file.write('](' + folder + '/' + file_name +')\n')
    index_file.write('\n')
index_file.close()
```
